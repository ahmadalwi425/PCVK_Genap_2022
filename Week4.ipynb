```python
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
import matplotlib.pyplot as plt
import numpy as np
```

```python
print('Change gamma of image')
print('--------------------------------')
try:
  inpGamma = int(input('Input gamma level: '))
except ValueError:
  print('Error, not a number')

original = cv.imread('/content/drive/MyDrive/kampus/PCVK/female.tiff')
gamma_image = np.zeros(original.shape, original.dtype)

#access for each pixel
for y in range(original.shape[0]):
  for x in range(original.shape[1]):
    for c in range(original.shape[2]):
      gamma_image[y,x,c] = ((original[y,x,c]/255)**(1/inpGamma)) * 255

final_frame = cv.hconcat((original, gamma_image))
cv2_imshow(final_frame)
```

```python
print('Change Depth of image')
print('--------------------------------')
try:
  inpDepth = int(input('Input Depth level: '))
except ValueError:
  print('Error, not a number')

original = cv.imread('/content/drive/MyDrive/kampus/PCVK/female.tiff')
Depth_image = np.zeros(original.shape, original.dtype)

#access for each pixel
for y in range(original.shape[0]):
  for x in range(original.shape[1]):
    for c in range(original.shape[2]):
      level = 255/((2**inpDepth)-1)
      Depth_image[y,x,c] = (round((original[y,x,c])/level)*level)

final_frame = cv.hconcat((original, Depth_image))
cv2_imshow(final_frame)
```

```python
print('Masked image "OR"')
print('--------------------------------')

original = cv.imread('/content/drive/MyDrive/kampus/PCVK/teeth.jpg')
blank_image = np.zeros(original.shape, original.dtype)
cv.rectangle(blank_image, pt1=(400, 25), pt2=(850,450), color=(255,255,255), thickness=-1)
cv.rectangle(blank_image, pt1=(310, 600), pt2=(780,1000), color=(255,255,255), thickness=-1)

masked_image = original | blank_image

final_frame = cv.hconcat((original, masked_image))
cv2_imshow(final_frame)
```

```python
print('Masked image "NOT"')
print('--------------------------------')

original = cv.imread('/content/drive/MyDrive/kampus/PCVK/teeth.jpg')
blank_image = np.zeros(original.shape, original.dtype)
cv.rectangle(blank_image, pt1=(400, 25), pt2=(850,450), color=(255,255,255), thickness=-1)
cv.rectangle(blank_image, pt1=(310, 600), pt2=(780,1000), color=(255,255,255), thickness=-1)

masked_image = ~original

final_frame = cv.hconcat((original, masked_image))
cv2_imshow(final_frame)
```

```python
print('Masked image "NAND"')
print('--------------------------------')

original = cv.imread('/content/drive/MyDrive/kampus/PCVK/teeth.jpg')
blank_image = np.zeros(original.shape, original.dtype)
cv.rectangle(blank_image, pt1=(400, 25), pt2=(850,450), color=(255,255,255), thickness=-1)
cv.rectangle(blank_image, pt1=(310, 600), pt2=(780,1000), color=(255,255,255), thickness=-1)

masked_image = ~(original & blank_image)

final_frame = cv.hconcat((original, masked_image))
cv2_imshow(final_frame)
```
